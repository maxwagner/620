### Make a bipartite network out of the movies and user, and then connect them with the ratings.

```python
G = nx.Graph()
```

```python
G.add_nodes_from(db.user.unique(), bipartite=0)
```

```python
G.add_nodes_from(db.movie.unique(), bipartite=1)
```

```python
weighting = [zip(db["user"], db["movie"], db["rating"])]
for edge in weighting:
    G.add_weighted_edges_from(edge)
```

```python
movies, users = bipartite.sets(G)
```

### Verify that the graph has the right number of nodes and edges, and take a look at the average degree.

```python
print(nx.info(G))
```

### Island Method Approach (For Users)

```python
def trim_edges(g, weight = 1):
    g2 = nx.Graph()
    for f, to, edata in g.edges(data=True):
        if edata['weight'] > weight:
            g2.add_edge(f,to,edata)
    return g2

def island_method(g, iterations = 5):
    weights = [edata['weight'] for f,to,edata in g.edges(data=True)]
    mn=int(min(weights))
    mx=int(max(weights))
    step = int((mx-mn)/iterations)
    return [[threshold, trim_edges(g, threshold)] for threshold in range(mn,mx,step)]
```

```python
user_ccs = list(nx.connected_component_subgraphs(G))[0]
```

```python
islands = island_method(user_ccs, iterations = 4) # 5 gives an error
for island in islands:
    print island[0], len(island[1]), len(list(nx.connected_component_subgraphs(island[1])))
```

### Let's compare the original graph to the modified version from the island method. The first graph is the full network, with all the nodes and edges. The second graph is modified with the island method at the 2nd threshold, the third graph is modified with the island method at the 4th threshold.

```python
plt.figure(figsize = (13, 13))
plt.axis('off')
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, nodelist=movies, node_color='m', alpha = .2)
nx.draw_networkx_nodes(G, pos, nodelist=users, node_color='c', alpha = .2)
nx.draw_networkx_edges(G, pos, alpha = .2)
```

```python
island_graph = trim_edges(user_ccs, 2)
movies, users = bipartite.sets(island_graph)

plt.figure(figsize = (13, 13))
plt.axis('off')
pos=nx.spring_layout(island_graph)
nx.draw_networkx_nodes(island_graph, pos, nodelist=movies, node_color='m', alpha = .2)
nx.draw_networkx_nodes(island_graph, pos, nodelist=users, node_color='c', alpha = .2)
nx.draw_networkx_edges(island_graph, pos, alpha = .2)
```

```python
island_graph = trim_edges(user_ccs, 4)
movies, users = bipartite.sets(island_graph)

plt.figure(figsize = (13, 13))
plt.axis('off')
pos=nx.spring_layout(island_graph)
nx.draw_networkx_nodes(island_graph, pos, nodelist=movies, node_color='m', alpha = .2)
nx.draw_networkx_nodes(island_graph, pos, nodelist=users, node_color='c', alpha = .2)
nx.draw_networkx_edges(island_graph, pos, alpha = .2)
```
