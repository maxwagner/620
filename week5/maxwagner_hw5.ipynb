```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import Image
```

---

### Overview

```python
G = nx.read_gml('lesmis.gml')
nx.draw(G)
```

```python
nx.number_of_nodes(G) # number of nodes
```

---

### Degree Centrality

```python
plt.hist(nx.degree(G).values())
plt.title('Degree Centrality')
plt.show()
```

---

### Eigenvector Centrality

```python
plt.hist(nx.eigenvector_centrality(G).values())
plt.title('Eigenvector Centrality')
plt.show()
```

### Conclusions

A few things can be concluded from the above. The most obvious is that both measurements are at their highest when both males and females are included. The eigenvector centrality for both genders vs one or the other is nearly double. The difference for degree centrality is less severe. 

The difference between males and females was less than expected. For eigenvector centrality both gender were essentially the same. Degree centrality had a one connection difference, which could be influenced by the amount of males compared to females, as degree centrality is a pure number of connections measurement, and not weighted.
