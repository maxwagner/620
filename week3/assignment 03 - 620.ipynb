```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
G = nx.read_gml('dolphins.gml')
nx.draw(G)
```

We can see that the above isn't too crazy, I don't think we will need to cut out any sections for the analysis.

Below is the diameter of the graph and a few other basic things.

```python
nx.diameter(G) # diameter
```

```python
nx.number_of_nodes(G) # number of nodes
```

A nice aspect of a network of this format (a social one), is that it forms a sort of hierarchy. If person A wants to meet person Z, but only knows person B, a calculation can be done in order to find who they can talk to, in which order, to get to person Z. For example below:

```python
nx.shortest_path(G, 0, 61)
```

```python
nx.shortest_path(G, 5, 61)
```

It is also possible to find the average number of connections needed for the entire network.

```python
nx.average_shortest_path_length(G)
```

Degree of connectivity can also be found.

```python
nx.average_degree_connectivity(G) # method 1 of getting average degree of connectivity
```

```python
nx.k_nearest_neighbors(G) # method 2 of getting average degree of connectivity
```
